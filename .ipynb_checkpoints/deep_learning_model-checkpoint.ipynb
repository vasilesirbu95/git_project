```python
path_data_csv = "cleaned_autoscout24-germany-dataset.csv"
```

```python
#def deep_learning_model(path_data_csv):
    
    # Matplotlib config
%matplotlib inline
%config InlineBackend.figure_formats = ['svg']
%config InlineBackend.rc = {'figure.figsize': (5.0, 3.0)}

import numpy as np
import pandas as pd
import seaborn as sns
```

```python
dataset = pd.read_csv(path_data_csv)
dataset.head()
```

```python
data_one_car_model = dataset[(dataset["Marke"] == "Jaguar") & (dataset["Model"] == "XF")]
print(data_one_car_model)
```

```python
#X = data_one_car_model[["Baujahr", "Kilometerstand", "Leistung", "Getriebe", "Kraftstoff"]]
X = data_one_car_model[["Baujahr", "Kilometerstand", "Leistung"]]
y = data_one_car_model["Preis"]
```

```python
# Tensorflow laden
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

model = keras.Sequential([
    keras.Input(shape = (5,)),
    layers.Dense(128, name = "hidden1", activation = "relu"),
    layers.Dense(1, name = "neuron", activation = "sigmoid")
])

print(model.summary())
```

```python
model.compile(
    optimizer = keras.optimizers.RMSprop(0.01),
    loss = keras.losses.BinaryCrossentropy()
)

model.fit(X.astype(np.float32), y, batch_size = 64, epochs = 100)
```

```python
np.mean((model.predict(X) > 0.5).ravel() == y)
```
