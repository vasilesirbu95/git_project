## Entscheidungsbaumregression

### 1. Hyperparameter zur Steuerung der Baumtiefe:

- #### max_depth:
Dies ist der wichtigste Hyperparameter, der die maximale Tiefe des Baums begrenzt.
Eine höhere Tiefe kann zu Überanpassung führen, während eine niedrigere Tiefe zu Unteranpassung führen kann.
Die optimale Tiefe hängt von der Komplexität der Daten ab.

- #### min_samples_split:
Dies ist die Mindestanzahl von Samples, die für die Teilung eines internen Knotens erforderlich sind.
Ein höherer Wert verhindert, dass der Baum zu spezifisch für die Trainingsdaten wird.

- #### min_samples_leaf:
Dies ist die Mindestanzahl von Samples, die in einem Blattknoten vorhanden sein müssen.
Ähnlich wie min_samples_split hilft dies, Überanpassung zu vermeiden.

- #### max_leaf_nodes:
Begrenzt die maximale Anzahl von Blattknoten.
min_impurity_decrease:
Ein Knoten wird nur aufgespalten, wenn diese Aufspaltung zu einer Verringerung der Ungenauigkeit(Impurity) führt, die größer oder gleich diesem Wert ist.

### 2. Hyperparameter zur Steuerung der Feature-Auswahl:

- #### max_features:
Dies ist die Anzahl der Features, die bei der Suche nach der besten Teilung berücksichtigt werden sollen.
Ein niedrigerer Wert kann die Vielfalt des Baums erhöhen und Überanpassung reduzieren.

### 3. Hyperparameter zur Steuerung des Wachstums des Baumes:

- #### criterion:
Dies ist die Funktion, die zur Messung der Qualität einer Teilung verwendet wird.
Für Regressionsbäume sind gängige Kriterien "mse" (mittlerer quadratischer Fehler) und "mae" (mittlerer absoluter Fehler).
Für Klassifikationsbäume sind gängige Kriterien „gini“ oder „entropy“.

- #### random_state:
Kontrolliert die Zufälligkeit des Schätzers.


```python

# Trainieren des Modells
model = DecisionTreeRegressor(min_samples_split = 16, min_samples_leaf = 3, max_features = 2, random_state=42)

# Hyperparameter-Grid erstellen
param_grid = {
    'max_depth': [7, 8, 9],
    'min_samples_split': [9, 10, 11],
    'min_samples_leaf': [1, 2, 3],
    'max_features': [2, 3, 4]
}

# GridSearchCV initialisieren
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=3, n_jobs=-1, verbose=2, scoring='r2')

# GridSearchCV trainieren
grid_search.fit(X_train, y_train)

# Beste Hyperparameter ausgeben
print("Beste Hyperparameter:", grid_search.best_params_)

# Bestes Modell aus GridSearchCV abrufen
best_rf_model = grid_search.best_estimator_

# Vorhersagen mit dem besten Modell
y_pred = best_rf_model.predict(X_test)
y_pred_train = best_rf_model.predict(X_train)

# R2-Werte ausgeben
print(f"R2_test für bestes RandomForest-Modell: {r2_score(y_test, y_pred)}")
print(f"R2_train für bestes RandomForest-Modell: {r2_score(y_train, y_pred_train)}")

# Erstellen der Scatterplots mit Linien für jede Methode
name = 'Entscheidungsbaumregression'
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, alpha=0.5)
plt.plot(y_test, y_test, color='red', linestyle='--')  # Linie für perfekte Vorhersagen
plt.title(f'Tatsächlicher Preis vs. Vorhergesagter Preis ({name})')
plt.xlabel('Tatsächlicher Preis')
plt.ylabel('Vorhergesagter Preis')
plt.grid(True)
plt.show()
```

## Random Forest Regression

### 1. Hyperparameter zur Steuerung der Anzahl und Größe der Bäume:

- #### n_estimators:
Dies ist die Anzahl der Bäume im Wald.
Eine höhere Anzahl von Bäumen führt in der Regel zu einer besseren Leistung, kann aber auch die Rechenzeit erhöhen.

- #### max_depth:
Die maximale Tiefe jedes einzelnen Baums.
Wie beim einzelnen Entscheidungsbaum kann eine höhere Tiefe zu Überanpassung führen.

- #### min_samples_split:
Die Mindestanzahl von Samples, die für die Teilung eines internen Knotens in jedem Baum erforderlich sind.

- #### min_samples_leaf:
Die Mindestanzahl von Samples, die in einem Blattknoten in jedem Baum vorhanden sein müssen.

- #### max_leaf_nodes:
Begrenzt die maximale Anzahl von Blattknoten.

### 2. Hyperparameter zur Steuerung der Feature-Auswahl:

- #### max_features:
Die Anzahl der Features, die bei der Suche nach der besten Teilung in jedem Baum berücksichtigt werden sollen.
Dies steuert die Vielfalt der Bäume und kann die Überanpassung reduzieren.

- #### bootstrap:
Bestimmt, ob Bootstrap-Samples verwendet werden sollen, wenn Bäume erstellt werden.

### 3. Weitere wichtige Hyperparameter:

- ### random_state:
Steuert die Zufälligkeit des Schätzers für die Reproduzierbarkeit.

- ### criterion:
Die Funktion, die verwendet wird um die Qualität eines Splits zu messen.
Für Regressionsbäume sind gängige Kriterien "mse" (mittlerer quadratischer Fehler) und "mae" (mittlerer absoluter Fehler).

- ### n_jobs:
Die Anzahl der Jobs, die parallel ausgeführt werden sollen.

```python

# Modell initialisieren
rf_model = RandomForestRegressor(random_state=42)

# Hyperparameter-Grid erstellen
param_grid = {
    'n_estimators': [200],
    'max_depth': [8],
    'min_samples_split': [3],
    'min_samples_leaf': [1],
    'max_features': ['sqrt']
}

# GridSearchCV initialisieren
grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, cv=3, n_jobs=-1, verbose=2, scoring='r2')

# GridSearchCV trainieren
grid_search.fit(X_train, y_train)

# Beste Hyperparameter ausgeben
print("Beste Hyperparameter:", grid_search.best_params_)

# Bestes Modell aus GridSearchCV abrufen
best_rf_model = grid_search.best_estimator_

# Vorhersagen mit dem besten Modell
y_pred = best_rf_model.predict(X_test)
y_pred_train = best_rf_model.predict(X_train)

# R2-Werte ausgeben
print(f"R2_test für bestes RandomForest-Modell: {r2_score(y_test, y_pred)}")
print(f"R2_train für bestes RandomForest-Modell: {r2_score(y_train, y_pred_train)}")

# Erstellen der Scatterplots mit Linien für jede Methode
name = 'Random Forest Regression'
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, alpha=0.5)
plt.plot(y_test, y_test, color='red', linestyle='--')  # Linie für perfekte Vorhersagen
plt.title(f'Tatsächlicher Preis vs. Vorhergesagter Preis ({name})')
plt.xlabel('Tatsächlicher Preis')
plt.ylabel('Vorhergesagter Preis')
plt.grid(True)
plt.show()
```

## Deep Learning

```python

# Skalieren der Merkmale
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Erstellen des Deep-Learning-Modells
model = keras.Sequential([
    layers.Dense(64, activation='relu', input_shape=(X_train_scaled.shape[1],)),
    layers.Dense(64, activation='relu'),
    layers.Dense(64, activation='relu'),
    layers.Dense(64, activation='relu'),
    layers.Dense(1)  # Ausgabeschicht für die Preisvorhersage
])

# Kompilieren des Modells
model.compile(optimizer='adam', loss='mse')

# Trainieren des Modells
model.fit(X_train_scaled, y_train, epochs=500, batch_size=4, verbose=0)

# Vorhersagen auf den Testdaten
y_pred = model.predict(X_test_scaled)
y_pred_train = model.predict(X_train_scaled)

# Berechnen des R²-Werts
# Berechnen des Bestimmungsmaßes R^2
print(f"R2_test Deep Learning: {r2_score(y_test, y_pred)}")
print(f"R2_train Deep Learning: {r2_score(y_train, y_pred_train)}")

# Erstellen der Scatterplots mit Linien für jede Methode
name = 'Deep Learning'
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, alpha=0.5)
plt.plot(y_test, y_test, color='red', linestyle='--')  # Linie für perfekte Vorhersagen
plt.title(f'Tatsächlicher Preis vs. Vorhergesagter Preis ({name})')
plt.xlabel('Tatsächlicher Preis')
plt.ylabel('Vorhergesagter Preis')
plt.grid(True)
plt.show()
```
